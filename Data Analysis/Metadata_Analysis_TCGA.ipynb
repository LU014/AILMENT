```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
metadata = pd.read_csv('Metadata-TCGA-All-18116-Samples.csv', index_col=0)
metadata = metadata[metadata['investigation'] == 'TCGA-COAD'] # Extract only colon cancer
metadata = pd.DataFrame(metadata)
metadata.index.name = 'patient_id'
pid_meta = metadata.index.astype('str')
metadata
```

```python
metadata.columns
```

```python
gender_mapping = {
    'MALE': 'Male',
    'FEMALE': 'Female',
}

ST_mapping = {
    'Metastatic': 'Others',
    'Recurrent Tumor': 'Others',
}

metadata['gender'] = metadata['gender'].replace(gender_mapping)
metadata['sample_type'] = metadata['sample_type'].replace(ST_mapping)
```

```python
gender_counts = metadata['gender'].value_counts()

labels = gender_counts.index
sizes = gender_counts.values
colors = ['olivedrab','orange','steelblue']

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.show()
```

```python
Sampletype_counts = metadata['sample_type'].value_counts()

labels = Sampletype_counts.index
sizes = Sampletype_counts.values
colors = ['olivedrab','orange','steelblue', 'rosybrown']

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.show()
```
